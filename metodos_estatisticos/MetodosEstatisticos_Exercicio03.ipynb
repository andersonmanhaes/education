1. Suponha que uma urna contenha 8 bolas vermelhas e 4 bolas brancas. Nós retiramos 2 bolas da urna sem reposição. Se nós assumimos que cada bola tem a mesma probabilidade de ser retirada, qual é a probabilidade de que ambas as bolas retiradas sejam vermelhas?

```python
#Quantidades
vermelhas = 8                #Quantidade de bolas vermelhas
brancas = 4                  #Quantidade de bolas brancas
total = vermelhas + brancas  #Total de bolas

print('Quantidades iniciais')
print(f'Bolas vermelhas: {vermelhas}\nBolas brancas: {brancas}\nTotasl: {total}')
```

```python
#Probabilidade inicial
prob_v = vermelhas / total
prob_b = brancas / total

print(f'P(vermelha) = {prob_v:.4f}\nP(branca) = {prob_b:.4f}')
```

```python
#Retirando a primeira bola vermelha
retira_vermelha = 1
vermelhas -= 1
total = vermelhas + brancas

print('Quantidades após retirando a primeira bola vermelha')
print(f'Bolas vermelhas: {vermelhas}\nBolas brancas: {brancas}\nTotasl: {total}')
```

```python
#Probabilidade após retirada a primeira bola
prob_rod2_v = vermelhas / total
prob_rod2_b = brancas / total

print(f'P(vermelha) = {prob_rod2_v:.3f}\nP(branca) = {prob_rod2_b:.3f}')
```

####**Resposta 1:**

```python
#Probabilidade das duas bolas serem vermelhas
prob_2_vermelhas = prob_v * prob_rod2_v

print(f'Probabilidade de ambas as bolas retiradas serem vermelhas: {prob_2_vermelhas:.3f}')
```

2. Um dado com seis lados apresenta quatro faces verdes (G) e duas vermelhas (R)  Ele é não viciado, de modo que cada face é equiprovável. O dado é então lançado algumas vezes. Calcule a probabilidade de ocorrência de cada uma destas sequências. Você deve escolher uma, dentre as seguintes três sequências de cores, de modo a maximizar a probabilidade de que nos primeiros lançamentos do dado ele gere exatamente a sequência que você escolheu. Qual destas sequências você escolheria? Justifique sua escolha.

> R G R R R

> R G R R R G

> G R R R R R

```python
#Quantidades
r = 2      #Quantidade de faces vermelhas
g = 4      #Quantidade de faces verdes
D = r + g  #Total de faces do dado

print('Características do dado')
print(f'Faces vermelhas: {r}\nFaces verdes: {g}\nTotal de faces: {D}')
```

```python
#Probabilidade individual de cada cor de face
prob_r = r / D
prob_g = g / D

print(f'P(r) = {prob_r:.3f}\nP(g) = {prob_g:.3f}')
```

```python
#Probabilidade da sequência 1: R G R R R
p1 = prob_r * prob_g * prob_r * prob_r * prob_r
p1
```

```python
#Probabilidade da sequência 2: R G R R R G
p2 = prob_r * prob_g * prob_r * prob_r * prob_r * prob_g
p2
```

```python
#Probabilidade da sequência 3: G R R R R R
p3 = prob_g * prob_r * prob_r * prob_r * prob_r * prob_r
p3
```

```python
#Probabilidade de cada sequência
print(f'P(S1) = {p1:.5f}\nP(S2) = {p2:.5f}\nP(S3) = {p3:.5f}')
```

3. Uma prova consiste de questões de múltipla escolha, cada qual apresentando 5 respostas possíveis. Suponha que você seja um estudante fazendo a prova, e que você estime que tem uma probabilidade de 0,75 de saber a resposta de qualquer questão que venha a ser perguntada e que, se você não sabe a resposta, você pretende chutar uma opção com probabilidade de 1/5 de estar correto. Qual é a probabilidade de você escolher a resposta correta em uma dada questão?

####**Resposta 3:**

Eventos:
* $A$: O aluno sabe a resposta
* $B$: A resposta está correta

Considerando os eventos $A$ e $B$, temos:

$P(A \cap B) = P(A)P(B|A) = 0.75 \times 1 = 0.75$

$P(A^c)P(B|A^c) = 0.25 \times 0.20 = 0.05$

Então a probabilidade do aluno acertar uma questão qualquer será:

$P(B) = P(A \cap B) + P(A^c)P(B|A^c) = 0.75 + 0.05 = 0.80$

4. Usando o terceiro axioma da probabilidade, prove que $P(A \cup B) = P(A) + P(B) - P(A \cap B)$ $^{(1)}$

####**Resposta 4:**
Através da decomposição das expressões, a  $P(A \cup B)$ pode ser escrita com a união de dois eventos mutamente exclusivos, $A \cup B = A \cup A^cB$, assim:

$P(A \cup B) = P(A) + P(A^cB)$ $^{(2)}$

Uma vez que $B = A^cB \cup (A \cap B$), temos:

$P(B) = P(A^cB) + P(A \cap B)$ $^{(3)}$

Removendo o valor de $P(A^cB)$ e substituindo na expressão $(2)$, temos:
$P(A \cup B) = P(A) + P(B) - P(A \cap B)$

Podemos interpretar estas afirmações de acordo com o terceiro axioma da probabilidade da seguinte maneira:

Para termos a probabilidade de que ao menos $A$ ou $B$ ocorram e que ambos não ocorram simultaneamente, teremos a probabilidade dada pelo somatório das chances dos eventos individuais e complementar que a inclusão do termo que subtrai a probabilidade da intersecção entre $A$ e $B$ na expressão $(1)$, garante que teremos apenas a soma de suas probabilidades na condição de eventos mutuamente exclusivos.
