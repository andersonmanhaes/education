1. Vinte adultos, cujas idades se encontram entre 30 e 40 anos, participaram de um estudo para avaliar o efeito de um regime específico de saúde envolvendo dieta e exercícios no colesterol. Dez desses adultos foram aleatoriamente assinalados a um grupo de controle, e outros 10 foram assinalados para o regime, como o grupo de tratamento, por um período de 6 meses. Os seguintes dados mostram a redução no colesterol experimentada ao longo deste período de tempo pelos 20 indivíduos:

> Grupo de controle: 7 3 -4 14 2 5 22 -7 9 5

> Grupo de tratamento: -6 5 9 4 4 12 37 5 3 3

(a) Compute a média, a mediana e a média truncada de 10% (trimmed mean) para ambos os grupos.

```python
import pandas as pd

controle = [7, 3, -4, 14, 2, 5, 22, -7, 9, 5]
tratamento = [-6, 5, 9, 4, 4, 12, 37, 5, 3, 3]

df_controle = pd.DataFrame(controle, columns = ['ReducaoColesterol_controle'])
controle_ordenado = df_controle.sort_values(['ReducaoColesterol_controle'])

df_tratamento = pd.DataFrame(tratamento, columns = ['ReducaoColesterol_tratamento'])
tratamento_ordenado = df_tratamento.sort_values(['ReducaoColesterol_tratamento'])

print('Grupo de Controle: \n')
print('Não ordenado:')
print(df_controle, '\n')
print('Ordenado:')
print(controle_ordenado, '\n')

print('Grupo de Tratamento: \n')
print('Não ordenado:')
print(df_tratamento, '\n')
print('Ordenado:')
print(tratamento_ordenado, '\n')
```



*   Média



```python
mean_df_controle = df_controle['ReducaoColesterol_controle'].mean()
mean_df_tratamento = df_tratamento['ReducaoColesterol_tratamento'].mean()

print('Média grupo de controle: ', mean_df_controle)
print('Média grupo de tratamento: ',  mean_df_tratamento)
```



*   Mediana





```python
median_df_controle = df_controle['ReducaoColesterol_controle'].median()
median_df_tratamento = df_tratamento['ReducaoColesterol_tratamento'].median()

print('Mediana grupo de controle: ', median_df_controle)
print('Mediana grupo de tratamento: ',  median_df_tratamento)
```



*   Média truncada de 10% (trimmed mean)



```python
tam_controle = len(controle_ordenado)
tam_tratamento_ordenado = len(tratamento_ordenado)
trimmed = 10
trimmed_persent = (trimmed/100)

count_trunca_controle = int(len(controle_ordenado)*trimmed_persent)
count_trunca_tratamento = int(len(tratamento_ordenado)*trimmed_persent)

df_controle_truncada = controle_ordenado[count_trunca_controle:-count_trunca_controle]
df_tratamento_truncada = tratamento_ordenado[count_trunca_tratamento:-count_trunca_tratamento]

media_df_controle_truncada = df_controle_truncada['ReducaoColesterol_controle'].mean()
media_df_tratamento_truncada = df_tratamento_truncada['ReducaoColesterol_tratamento'].mean()
'''
print(count_trunca_controle)
print(count_trunca_tratamento)
print(df_controle_truncada)
print(df_tratamento_truncada)
'''

print('Mediana grupo de controle truncada em 10%: ', media_df_controle_truncada)
print('Mediana grupo de tratamento truncada em 10%: ', media_df_tratamento_truncada)
```

(b) Explique o porquê a diferença nas médias sugere uma conclusão acerca do efeito do regime, enquanto que a diferença nas medianas ou na média truncada sugere uma conclusão distinta.

**Resposta:** Devido a existência de um valor díspar dos demais na amostra do grupo de tratamento ("outlier") e no cálculo da média este valor é utilizado, já na mediana e média truncada este valor não é sensibilizado no cálculo, podendo-se chegar a conclusões distintas.

```python
downtime = [22, 18, 135, 15, 90, 78, 69, 98, 102, 83, 55, 28, 121, 120, 13, 22, 124, 112, 70, 66, 74, 89, 103, 24, 21, 112, 21, 40, 98, 87, 132, 115, 21, 28, 43, 37, 50, 96, 118, 158, 74, 78, 83, 93, 95]

df_downtime = pd.DataFrame(downtime, columns = ['downtime'])
df_downtime_order = df_downtime.sort_values(['downtime'])

'''
print(df_downtime)
print(df_downtime_order)
'''

media_downtime = df_downtime['downtime'].mean()
mediana_downtime = df_downtime_order['downtime'].median()

print('Média amostral das durações de falha de energia: ', media_downtime)
print('Mediana amostral das durações de falha de energia: ', mediana_downtime)
```

>6,72 6,77 6,82 6,70 6,78 6,70 6,62 6,75

>6,66 6,66 6,64 6,76 6,73 6,80 6,72 6,76

>6,76 6,68 6,66 6,62 6,72 6,76 6,70 6,78

>6,76 6,67 6,70 6,72 6,74 6,81 6,79 6,78

>6,66 6,76 6,76 6,72

(a) Calcule a média amostral e o desvio padrão amostral.

```python
diametro = [6.72, 6.77, 6.82, 6.70, 6.78, 6.70, 6.62, 6.75, 6.66, 6.66, 6.64, 6.76, 6.73, 6.80, 6.72, 6.76, 6.76, 6.68, 6.66, 6.62, 6.72, 6.76, 6.70, 6.78, 6.76, 6.67, 6.70, 6.72, 6.74, 6.81, 6.79, 6.78, 6.66, 6.76, 6.76, 6.72]

df_diametro = pd.DataFrame(diametro, columns = ['diametro_cabeca_arrebite'])
media_diametro = df_diametro['diametro_cabeca_arrebite'].mean()
desvio_diametro = df_diametro['diametro_cabeca_arrebite'].std()

#print(df_diametro)
print('Média amostral dos diâmetros das cabeças de rebites: ', media_diametro)
print('Média amostral dos diâmetros das cabeças de rebites: ', desvio_diametro)
```

(b) Construa histogramas de frequência relativa. Discuta o impacto que a escolha da largura dos intervalos utilizados tem na informação visual disponibilizada pelos histogramas.

Resposta: A largura do intervalo impacta na quantidade de dados, neste caso os diametros das cabeças dos arrebites, que serão agrupados em um mesmo intervalo de frequência.

```python
import matplotlib.pyplot as plt
import math as mt

frequancia = int(mt.sqrt(len(diametro)))
plt.hist(diametro, frequancia, rwidth=0.9)
plt.xlabel('Diametro cabeça arrebite')
plt.ylabel('Qtd')
plt.plot()
plt.show()
```
